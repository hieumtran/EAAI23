```python
# train = pd.read_csv("archive/trainval_name_reglabel.csv")
train = pd.read_csv('train_resample_120000_reg.csv')
val_train = train.iloc[:, 1]
ars_train = train.iloc[:, 2]

test = pd.read_csv("archive/test_name_reglabel.csv")
val_test = test.iloc[:, 1]
ars_test = test.iloc[:, 2]
```

```python
def viz(data, viz_name):
    plt.figure(figsize=(15,7))
    plt.hist(data)
    plt.ylabel('Number of observation')
    plt.title(viz_name)
    plt.show()

viz(val_train, 'Valence train')
viz(ars_train, 'Arousal train')
```

```python
viz(val_train, 'Valence train')
viz(ars_train, 'Arousal train')
```

```python
viz(val_test, 'Valence test')
viz(ars_test, 'Arousal test')
```
