```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("archive/trainval_name_reglabel.csv")
# train = pd.read_csv('train_CJ_HF_20000_reg.csv')
val_train = train.iloc[:, 1]
ars_train = train.iloc[:, 2]

test = pd.read_csv("archive/test_name_reglabel.csv")
val_test = test.iloc[:, 1]
ars_test = test.iloc[:, 2]
```

```python
def histogram(x, title, save_file_name) :
    plt.figure()
    plt.hist(x)
    plt.title(title)
    plt.savefig(save_file_name)

histogram(val_train, 'valence', 'valence_img.png')
histogram(ars_train, 'arousal', 'arousal_img.png')
```

```python
def heatmap(x, y, bins=10, save_file_name='val_ars_distribution.png') : 
    plt.figure()
    plt.hist2d(x, y, bins=bins, cmap='plasma')

    cb = plt.colorbar()
    cb.set_label('Number of entries')

    plt.title('The distribution of emotion in the dataset')
    plt.xlabel('valence')
    plt.ylabel('arousal')

    plt.savefig(save_file_name)

heatmap(val_train, ars_train, bins=10, save_file_name='train_distribution.png')
heatmap(val_test, ars_test, bins=5, save_file_name='test_distribution.png')
```
